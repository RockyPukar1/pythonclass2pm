```python
class Hello:
    @staticmethod     #this rule should be used exaactly used just over the method that doesnot uses the properties of the clas
    def hello():      #This is called decorator
        print("Hello World")
        
obj=Hello()             #object
obj.hello() 
```

```python
class Volume:
    def __init__(self,l,b,h):
        self.l=l
        self.b=b
        self.h=h
        
    def volume(this):
        v=this.l*this.b*this.h
        print(v)
        
    def area(self):
        a=self.l*self.b
        print(a)
        
l=int(input("Enter l = "))
b=int(input("Enter b = "))
h=int(input("Enter h = "))
obj=Volume(l,b,h)  #When object is formed it automatically initializes __init__ function
obj.volume() 
obj.area()
print(obj.l,obj.b)
```

```python
class A:
    def __init__(self):
        self.name=input("Enter name= ")
        self.age=int(input("Enter age= "))
        
class B(A):
    def hello(self):
        print(f"Hello World I am {self.name}. I am {self.age} years old.")
        
obj=B()
obj.hello()
```

```python
#Multi level inheritance-->one inheritance at a time
class A:
    print("This is class A")
class B(A):
    print("This is class B")
class C(B):
    print("This is class C")

obj=C()
```

```python
#Multiple inheritance-->
class A:
    print("This is class A")
class B:
    print("This is class B")
class C(B,A):
    print("This is class C")

obj=C()
```

```python
class A:
    def __init__(self,add):
        self.add=add
        
    def my_add(self):
        return self.add
        
class B(A):
    def __init__(self,name,age,add):
        self.name=name
        self.age=age
        A.__init__(self,add) #This step to compulsory to send the (add) argument to class A
        
    def hello(self):
        print(f"Hello World I am {self.name}. I am {self.age} years old. I live in {self.add}.")
        

name=input("Enter name= ")
age=int(input("Enter age= "))
add=input("Enter add= ")

obj=B(name,age,add)
obj.hello()
address = obj.my_add()
print(f"My address is {address}")
```

```python
class A:
    def __init__(self,add):
        self.add=add
        
    def my_add(self):       # It is recommended to add at least one function
        return self.add
        
class B(A):
    def __init__(self,age,add):
        self.age=age
        A.__init__(self,add)
    def my_info(self):
        print(f"Hello World I am Ram. I am {self.age} years old. I am from {self.add}")

class C(B):
    def __init__(self,name,age,add):
        self.name=name
        B.__init__(self,age,add) #This step to compulsory to send the (add) argument to class A
        
    def hello(self):
        print(f"Hello World I am {self.name}. I am {self.age} years old. I live in {self.add}.")
        

name=input("Enter name= ")
age=int(input("Enter age= "))
add=input("Enter add= ")

obj=C(name,age,add)
obj.my_info()
obj.hello()
```

```python
#Public incapsulation
class Info():
    def __init__(self,name,age,add):
        self.name=name
        self.age=age
        self.add=add
        
obj=Info("Ram",23, "Kathmandu")
print(obj.name)
print(obj.age)
print(obj.add)
```

```python
#Protected incapsulation
class Info():
    def __init__(self,name,age,add):
        self._name=name
        self._age=age
        self._add=add
        
obj=Info("Ram",23,"kathmandu")
print(obj._name) #Protected properties can be acessed by using _(single underscore) anywhere
print(obj._age)
print(obj._add)
```

```python
#Private incapsulation
class Info():
    def __init__(self,name,age,add):
        self.__name=name
        self.__age=age
        self.__add=add
        
obj=Info("Ram",23,"kathmandu")
print(obj._Info__name)#Private properties can be only be acessed only inside same class inside other function.But we can also acess it other place by using (_class-name__).
                #This method is called name memmbling
print(obj._Info__age)
print(obj._Info__add)
```

```python
class A:
    def __init__(self):
        self.name=input("Enter name = ")
        self._age=int(input("Enter age = "))
        self.__add=input("Enter add = ")
        
    def a(self):
        print(f"Helloo World I am {self.name}. I am from {self.__add}. I am {self._age} years old")
        
class B(A):
    def hello(self):
        print(f"Hello World I am {self.name}. I am {self._age} years old.")
        
obj=B()
obj.hello()
obj.a()
```

```python
class Parrot:
    def fly(self):
        print("Parrot can fly")
    def swim(self):
        print("Parrot can't swim")
        
class Penguin:
    def fly(self):
        print("Penguin can't fly")
    def swim(self):
        print("Penguin can swim")
        
#common interface
def flying_test(bird):
    bird.fly()
def swimming_test(bird):
    bird.swim()
    
#instantiate objects
blu=Parrot()
peggy=Penguin()
#passing the object
flying_test(blu)
flying_test(peggy)
swimming_test(blu)
swimming_test(peggy)
```
